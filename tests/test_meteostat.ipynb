```python
from meteostat import Stations, Daily, Point, TimeSeries
from datetime import datetime
import pandas as pd
import altair as alt
```

## Get gps coord for a city, an address ...

```python
from geopy.geocoders import Nominatim

# Initialize the geocoder
geolocator = Nominatim(user_agent="my_geocoder")
```

```python
# Get coord for an address

address_list = [
    "Place de la Comédie, Montpellier",
    "361 rue du Pré aux Clercs, Montpellier"
]

# Initialize a list of geo dict
geo_list = []

# Get the coordinates for the address_list
for address in address_list:
    location = geolocator.geocode(address)

    if location:
        print(f"Coordinates for {address}: {location.latitude}, {location.longitude}, {location.altitude}")
        geo_list.append(location.raw)
    else:
        print(f"Could not find coordinates for {address}")
```

## Get nearby stations and analyze


```python
# convert (geo_list[0]['lat']) to real
geo_list[0]['lat'] = float(geo_list[0]['lat'])
geo_list[0]['lon'] = float(geo_list[0]['lon'])
```

```python
# Get nearby weather stations
stations = Stations()
stations = stations.nearby(geo_list[0]['lat'], geo_list[0]['lon'])
station_df = stations.fetch(50)

# Print DataFrame
station_df
```

```python
# plot stations in station_df in a folium map with tiles="cartodb positron", add stations as markers of color orange, show name in popup
import folium

m = folium.Map(location=[geo_list[0]['lat'], geo_list[0]['lon']], zoom_start=10, tiles="cartodb positron")

```

```python
for index, row in station_df.iterrows():
    folium.Marker([row['latitude'], row['longitude']],
                  tooltip=index + ' ' + row['name'],
                  icon=folium.Icon(color='orange')
                 ).add_to(m)

# Display the map
m
```

## get data for a point


```python

# Set time period
start = datetime(2025, 6, 15)
end = datetime(2025, 7, 31)

# Get daily data
data = Daily('07643', start, end)
data_df = data.fetch()

data_df
```

```python
data.stations
```

```python
# Plot tmin and tmax against time
chart_temp = alt.Chart(data_df.reset_index()).mark_line().encode(
    x='time:T',
    y=alt.Y('value:Q', title='Temperature (°C)'),
    color='variable:N'
).transform_fold(
    ['tmin', 'tmax'],
    as_=['variable', 'value']
).properties(
    title='Minimum and Maximum Temperature Over Time',
    height=300, # Set height for the temperature chart
    width=800   # Set width for the temperature chart
)

# Plot prcp against time
chart_prcp = alt.Chart(data_df.reset_index()).mark_bar(color='blue').encode(
    x='time:T',
    y=alt.Y('prcp:Q', title='Precipitation (mm)')
).properties(
    title='Precipitation Over Time',
    height=300, # Set height for the precipitation chart
    width=800   # Set width for the precipitation chart
)

# Combine the charts
combined_chart = alt.vconcat(chart_temp, chart_prcp)

# Display the combined chart
combined_chart
```

```python
# plot tmin for data_df.tmin et data1_df.tmin, on the same graph with Altair
# Combine the 'tmin' data from both dataframes
combined_tmin_df = pd.concat([data_df['tmin'].rename('data_df'), data1_df['tmin'].rename('data1_df')], axis=1)

# Plot tmin for both dataframes against time
chart_tmin_comparison = alt.Chart(combined_tmin_df.reset_index()).mark_line().encode(
    x='time:T',
    y=alt.Y('value:Q', title='Minimum Temperature (°C)'),
    color='variable:N'
).transform_fold(
    ['data1_df', 'data_df'],
    as_=['variable', 'value']
).properties(
    title='Minimum Temperature Comparison Over Time',
    height=600, # Set height for the temperature chart
    width=800   # Set width for the temperature chart
)

# Display the chart
chart_tmin_comparison
```
